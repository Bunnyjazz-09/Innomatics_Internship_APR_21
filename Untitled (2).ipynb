<h2>Research Question</h2>

<h4>A person claims that his average win is more than 6.05. Given that the population std is 7.3.</h4>

<p>Step-1)</p>
<p>Null Hypothesis:-</p>                  $$H_0 = 6.05$$
<p>Alternate Hypothesis :-</p>            $$H_1 > 6.06$$

<p>Step-2</p>
<p>Collect a sample of size n = 10</p>
    $$ [4.45, 3.2, 5.8, 4.9, 5.79, 3.08, 6.5, 7.8, 4.09, 2.03] $$
<p>Compute sample mean</p>

<p>Step-3</p>
<p>Compute Test Statistic:</p>
$$ z = \frac{\bar{x} - \mu}{\sigma/\sqrt[2]{n}}$$

<p>Step-4</p>
 <p>Decide $ \alpha $</p>

<p>Step-5</p>
<p>Apply left tailed z-test</p>
$ z &lt; z_{n-1,\alpha} \Rightarrow \mbox{Reject } H_0 \mbox{ or Accept }H_1$

```python
from scipy.stats import norm
```

```python
def z_score(sample_size, sample_mean, pop_mean, pop_std):
    numerator = sample_mean - pop_mean
    denomenator = pop_std / sample_size**0.5
    return numerator / denomenator
```

```python
confidence_level = 0.95

alpha = 1 - confidence_level

z_critical = norm.ppf(1 - alpha)

print(z_critical)
```

```python
sample_size = 10
sample_mean = sum([4.45,3.2,5.8,4.9,5.79,3.08,6.5,7.8,4.09,2.03])/10
sample_mean
pop_mean = 6.05
pop_std = 7.3
```

```python
z = z_score(sample_size, sample_mean, pop_mean, pop_std)

print(z)
```

```python

x_min = 2.0
x_max = 8.0

mean = pop_mean
std = pop_std / (sample_size**0.5)

x = np.linspace(x_min, x_max, 100)
y = norm.pdf(x, mean, std)
plt.xlim(x_min, x_max)
plt.plot(x, y)

z_critical_left = pop_mean + (-z_critical * std)

x1 = np.linspace(x_min,z_critical_left, 100)
y1 = norm.pdf(x1, mean, std)
plt.fill_between(x1, y1, color='orange')

plt.scatter(sample_mean, 0)
plt.annotate("x_bar", (sample_mean, 0.1))
```

```python
if(z < -z_critical):
    print("Reject Null Hypothesis")
else:
    print("Fail to reject Null Hypothesis")
```

```python
p_value = 1.0 - norm.cdf(np.abs(z))

print("p_value = ", p_value)

if(p_value < alpha):
    print("Reject Null Hypothesis")
else:
    print("Fail to reject Null Hypothesis")
```
